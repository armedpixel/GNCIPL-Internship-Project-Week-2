```python
df=df.fillna({'Date':'2015-01-02','Industry':"Other",'Location':'Other','Investment Type':'Other'})
```

```python
duplicate_values = df.duplicated().sum()
print(f"There are {duplicate_values} duplicate values present in the dataset.")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
cat = df.select_dtypes(['object'])
cat_columns = cat.columns

for col in cat_columns:
    print(col, df[col].unique())
    print(" ")
```

```python
def clean_location(value):
    if pd.isna(value):
        return 'Unknown / Other'
    
    value = value.lower().strip().replace('\xa0', '')

    # Mapping logic
    if any(loc in value for loc in ['bengaluru', 'bangalore', 'kormangala', 'sanfrancisco/bangalore', 'sfo/bangalore']):
        return 'Bengaluru'
    elif any(loc in value for loc in ['mumbai', 'andheri', 'chembur', 'mumbai/bengaluru', 'mumbai/global', 'mumbai/uk', 'mumbai/ny']):
        return 'Mumbai'
    elif any(loc in value for loc in ['delhi', 'newdelhi', 'gurgaon', 'gurugram', 'noida', 'nwdelhi']):
        return 'Delhi NCR'
    elif 'pune' in value:
        return 'Pune'
    elif 'hyderabad' in value:
        return 'Hyderabad'
    elif 'chennai' in value or 'taramani' in value:
        return 'Chennai'
    elif 'ahmedabad' in value or 'ahemadabad' in value or 'ahemdabad' in value:
        return 'Ahmedabad'
    elif 'kolkata' in value or 'kolkatta' in value:
        return 'Kolkata'
    elif any(loc in value for loc in ['jaipur', 'bhopal', 'indore', 'nagpur', 'surat', 'vadodara', 'trivandrum', 'coimbatore', 'goa', 'agra', 'udaipur', 'jodhpur', 'belgaum', 'hubli', 'kanpur', 'lucknow', 'varanasi', 'srinagar', 'chandigarh', 'bhubaneswar', 'bhubneswar', 'siliguri', 'gaya', 'gwalior', 'karur', 'udupi', 'amritsar']):
        return 'Other Indian Cities'
    elif any(loc in value for loc in ['paloalto', 'sanfrancisco', 'stanford', 'wilmington', 'newark', 'menlopark', 'santamonica', 'sfo', 'usa', 'us', 'seattle', 'boston', 'newyork']):
        return 'US Cities'
    elif 'singapore' in value:
        return 'Singapore'
    else:
        return 'Other'
df['City'] = df['Location'].apply(clean_location)

# Optional: check results
print(df['City'].value_counts())
```

1) Time series analysis 2015 to 2021

```python

df['Year']=df['Date'].dt.year
df.groupby('Year')['Amount in USD'].sum().plot.bar(title='Total Investment Amount by Year')
```

```python
df.groupby('Year')['Amount in USD'].count().plot.bar(title='Total Investment Number by Year')
```

```python
df.groupby('Year')['Amount in USD'].mean().plot.bar(title='Investment Average Amount by Year')
```

```python
df['Startup Name']=df['Startup Name'].str.lower()
df['Industry']=df['Industry'].str.lower()
df['Sub-vertical']=df['Sub-vertical'].str.lower()
df['Location']=df['Location'].str.lower()
df['Investors']=df['Investors'].str.lower()
df['Investment Type']=df['Investment Type'].str.lower()
df=df.replace(' ','',regex=True)
df=df.replace('-','',regex=True)
```

Top 15 industries which took lots of investment

```python
df.groupby('Industry')['Amount in USD'].sum().sort_values(ascending=False).head(15).plot.bar(figsize=(12,6), title='Total Funding by Industry')
plt.ylabel('Amount in USD')
```

Top 15 start ups which took lots of investment

```python
df.groupby('Startup Name')['Amount in USD'].sum().sort_values(ascending=False).head(15).plot.bar(figsize=(10,8),title='Total Funding by Start up')
plt.ylabel('Amount in USD')
```

Top 15 Investors

```python
df.groupby('Investors')['Amount in USD'].sum().sort_values(ascending=False).head(15).plot.bar(figsize=(10,8))
```

```python
funding = df.pivot_table(values='Amount in USD',
                         index='City',
                         columns='Industry',
                         aggfunc='sum')

top_cities = funding.sum(axis=1).sort_values(ascending=False).head(10).index
top_industries = funding.sum(axis=0).sort_values(ascending=False).head(10).index

filtered_data = funding.loc[top_cities, top_industries]

plt.figure(figsize=(16, 8))
sns.heatmap(filtered_data, annot=True, fmt='.0f', cmap='coolwarm', linewidths=0.5, linecolor='gray')
plt.title('Funding Amount by City and Industry')
plt.xlabel('Industry Vertical')
plt.ylabel('City Location')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
stage_funding = df.groupby('Investment Type')['Amount in USD'].sum().sort_values(ascending=False).head(20)

sns.barplot(x=stage_funding.values, y=stage_funding.index,hue=stage_funding.index , palette='magma')
plt.title('Stage-wise Funding Amounts')
plt.xlabel('Amount in USD')
plt.ylabel('Funding Stage')
plt.show()
```
